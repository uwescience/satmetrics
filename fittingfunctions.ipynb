```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as optim
```

```python
def gauss(x, a=1, mu=0, sigma=2):
    return a*np.exp(-(((x-mu)/(2*sigma))**2))
```

```python
x = np.arange(-10, 10, 0.5)
```

```python
y = gauss(x, 7, 5, 1)
```

```python
len(y)
```

```python
plt.scatter(x,y)
```

```python
noise = np.random.normal(size=(len(y)))/10
```

```python
realdata = y + noise
plt.plot(realdata)
```

```python
(a, mu, sigma), unc = optim.curve_fit(gauss, x, realdata)
```

```python
plt.plot(x, realdata)
plt.plot(x, gauss(x, a, mu, sigma))
```

```python
(a, mu, sigma), unc = optim.curve_fit(gauss, x, noise)
print((a, mu, sigma), unc)
```

```python
z = optim.curve_fit(gauss, x, noise, p0=[200, 300, 50], full_output=True)
print(z)
```

```python
#if its much less than 1 then our rmse is not good
np.sqrt(np.sum((noise - gauss(x, a, mu, sigma))**2)/len(x))
```
